정사각형의 점의 수 : 1/4원 안의 점의 수 = 정사각형의 넓이 : 1/4원의 넓이
n_point : n = r**2 : pi*r**2/4
            = 1 : pi/4
pi = 4*n/n_point

```
import numpy as np
import matplotlib.pyplot as plt

n_point = 10**4
n = 0
point = np.random.rand(n_point, 2)

plt.figure(figsize=(5,5))
for i in range(n_point):
  x, y = point[i][0], point[i][1]
  r = x**2 + y**2
  if r<1:
    color = 'r'
    n = n+1
  else:
    color = 'b'
  plt.scatter(x, y, s=0.2, c=color)
plt.xlim(0,1)
plt.ylim(0,1)
plt.show()

print(4*n/n_point)
```

```
import numpy as np
import matplotlib.pyplot as plt

count_list = []
pi_list = []

count = 10
while count < 10**7:
  n = 0
  point = np.random.rand(count, 2)
  for i in range(count):
    x, y = point[i][0], point[i][1]
    r = x**2 + y**2
    if r<1:
      n = n+1
  count_list.append(count)
  pi_list.append(4*n/count)
  print(count, 4*n/count)
  count = count*3

ind = np.arange(len(pi_list))
plt.plot(ind, pi_list)
plt.xticks(ind, count_list)
plt.xticks(rotation = 35)
plt.show()
```
